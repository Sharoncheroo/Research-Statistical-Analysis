```python
def period_to_date(data, column):

   data[column] = pd.to_datetime(data[column], format='%b-%y')
   return data

#data['period'] = pd.to_datetime(data['period']) # convert to date dtype
```

```python
def county_cleanup(data, column):
  data[column] = data[column].str.replace('County', "") # remove county redundancy
  data[column] = data[column].str.strip() # trim unnecessary spaces
  return data
```

```python
#call the functions
county_cleanup(data, 'county')
period_to_date(data, 'period')
data.info()
```

```python
def missing_values_ratios():

  percent_missing = list(data.isnull().sum()/len(data)*100)
  data_cols = data.columns
  missing_ratio = dict(map(lambda i, j : (i,j), data_cols, percent_missing))

  high_ratio = {}
  for key, value in missing_ratio.items():
    if value >= 50:
      high_ratio[key]=value

  low_ratio = {}
  for key, value in missing_ratio.items():
    if value < 50:
      high_ratio[key]=value
  
  return(high_ratio, low_ratio)
```

```python
def sum_columns_with_prefix_and_replace(data, prefix):
    filtered_columns = [col for col in data.columns if col.startswith(prefix)]
    new_column = data[filtered_columns].sum(axis=1)
    data.drop(columns=filtered_columns, inplace=True)  # Drop the original columns
    data[prefix] = new_column  # Add the new column
    return data

# call function
data = sum_columns_with_prefix_and_replace(data, 'stunted')
data = sum_columns_with_prefix_and_replace(data, 'Underweight')
```

# Descriptive statistics

```python
# Extracting year from the 'period' column
data['Year'] = pd.to_datetime(data['period']).dt.year

# Creating the table with sum of cases reported
statistics_table = data.groupby('Year').agg({
    'total dewormed': 'mean',
    'diarrhoea cases': 'mean',
    'stunted': 'mean',
    'underweight': 'mean'
}).reset_index()

# Convert sum columns to integers
sum_columns = ['total dewormed', 'diarrhoea cases', 'stunted', 'underweight']
statistics_table[sum_columns] = statistics_table[sum_columns].round()
statistics_table[sum_columns] = statistics_table[sum_columns].astype(int)

# Renaming the columns for clarity
statistics_table.columns = ['Year', 'Dewormed', 'Diarrhea', 'Stunted', 'Underweight']

# Displaying the statistics table
print(statistics_table)
```
